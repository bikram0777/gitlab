```python
x=np.arange(1,11)
y=np.arange(5,15)
x
y
```

```python
plt.figure(figsize=(2,2))
plt.plot(x,x*x*x,c='r',linestyle='dashed')
plt.xlabel("X-Axis")
plt.ylabel("y-Axis")
plt.title("X and Y Axis")
plt.legend("x",loc="upper right")
```

```python
plt.figure(figsize=(4,4))
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x,x*x)
plt.subplot(2,2,4)
plt.plot(x,x*x*x)
```
